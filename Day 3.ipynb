# **Matplotlib with Pandas**
### 1. Importing Libraries


### Key Points:


*   **kind parameter:** Specifies the type of plot. Options include 'line',
    'bar', 'hist', 'scatter', 'pie', 'box', and 'area'
*   **x and y parameters:** Specify which columns to use for the x and y axes.
*   **plt.show():** Displays the plot.





```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Creating a Simple DataFrame


```python
# Example DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 220, 250, 280, 300]
}
df = pd.DataFrame(data)
df.head()
```

### 3. Line Plot

```python
df.plot(x='Month', y='Sales')
plt.show()
```

```python
df.plot(x='Month', y='Sales', kind='line', marker='o', title='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### 4. Bar Plot

```python
df.plot(x='Month', y='Sales', kind='bar', color='skyblue', title='Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### 5. Histogram

```python
# Example DataFrame
data = {'Scores': [10, 92, 40, 79, 20, 95, 87, 67, 88, 75,5]}
df = pd.DataFrame(data)

df.plot(kind='hist', bins=10, edgecolor='black', title='Scores Distribution')
plt.xlabel('Scores')
plt.show()
```

### 6. Scatter Plot

```python
# Example DataFrame
data = {
    'Hours': [1, 2, 3, 4, 5],
    'Scores': [50, 60, 70, 80, 90]
}
df = pd.DataFrame(data)

df.plot(x='Hours', y='Scores', kind='scatter', color='red', title='Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.show()
```

### 7. Pie Chart

```python
# Example DataFrame
data = {'Category': ['A', 'B', 'C', 'D'],
        'Values': [10, 20, 30, 40]}
df = pd.DataFrame(data)

df.plot(kind='pie', y='Values', labels=df['Category'], autopct='%1.1f%%', title='Category Distribution')
plt.ylabel('')
plt.show()
```

### 8. Box Plot

```python
# Example DataFrame
data = {'Scores': [88, 92, 85, 79, 90, 95, 87, 91, 88, 84]}
df = pd.DataFrame(data)

df.plot(kind='box', title='Score Distribution')
plt.ylabel('Scores')
plt.show()
```

### 9. Area Plot

```python
# Example DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 220, 250, 280, 300]
}
df = pd.DataFrame(data)

df.plot(x='Month', y='Sales', kind='area', title='Monthly Sales Area Plot')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

# **Seaborn with Pandas Cheat Sheet**
### 1. Importing Libraries

### Key Points:


*   **data parameter:** Specifies the DataFrame to use.
*   **x and y parameters:** Define the columns for the x and y axes.
*   **hue parameter:** Allows you to add another dimension by coloring points
    by categories.
*   **plt.show():** Displays the plot.





### 2. Creating a Simple DataFrame

```python
# Example DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 220, 250, 280, 300]
}
df = pd.DataFrame(data)
```

### 3. Line Plot

```python
sns.lineplot(x='Month', y='Sales', data=df, marker='o')
```

```python
sns.lineplot(x='Month', y='Sales', data=df, marker='o')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### 4. Bar Plot

```python
sns.barplot(x='Month', y='Sales', data=df, color='red')
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

### 5. Histogram

```python

# Example DataFrame
data = {'Scores': [88, 92, 85, 79, 90, 95, 87, 91, 88, 84]}
df = pd.DataFrame(data)

sns.histplot(df['Scores'], bins=5, color='blue')
plt.title('Scores Distribution')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.show()
```

### 6. Scatter Plot

```python
# Example DataFrame
data = {
    'Hours': [1, 2, 3, 4, 5],
    'Scores': [50, 60, 70, 80, 90]
}
df = pd.DataFrame(data)

sns.scatterplot(x='Hours', y='Scores', data=df, color='red')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.show()
```

### 8. Box Plot

```python
# Example DataFrame
data = {'Scores': [-12,88, 92, 85, 79, 90, 95, 87, 91, 88, 84,130]}
df = pd.DataFrame(data)

sns.boxplot(y='Scores', data=df, color='lightgray')
plt.title('Score Distribution')
plt.ylabel('Scores')
plt.show()
```

```python
# Example DataFrame
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May','Jun','Sep'],
    'Sales': [-200, 220, 250, 280, 300,-12,200]
}
df = pd.DataFrame(data)
```
